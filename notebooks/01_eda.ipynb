```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Data Loading

```python
# Load the IMDb dataset
data_path = '../data/raw/IMDB Dataset.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the dataset
df.head()
```

### Basic Information

```python
# Get basic information about the dataset
df.info()

# Display the shape of the dataset
print(f"Dataset Shape: {df.shape}")
```

### Missing Values

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

### Sentiment Distribution

```python
# Visualize the distribution of sentiments
sns.countplot(x='sentiment', data=df, palette='Set2', hue='sentiment')
plt.title('Distribution of Positive and Negative Reviews')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.savefig('../reports/sentiments_dist.png')
plt.show()
```

### Text Length Analysis

```python
# Calculate the length of each review
df['review_length'] = df['review'].apply(lambda x: len(x.split()))

# Visualize the distribution of review lengths
plt.figure(figsize=(10, 6))
sns.histplot(df['review_length'], bins=50, kde=True)
plt.title('Distribution of Review Lengths')
plt.xlabel('Length of Review (Number of Words)')
plt.ylabel('Frequency')
plt.savefig('../reports/review_len.png')
plt.show()
```

### Conclusion

The Exploratory Data Analysis (EDA) phase provided valuable insights into the IMDb Movie Reviews dataset, which consists of 50,000 entries labeled as either positive or negative sentiments. Key findings from the analysis include:

- **Data Overview:** The dataset is well-structured, comprising two columns: review and sentiment, with no missing values present.

- **Sentiment Distribution:** The dataset is balanced, containing an equal number of positive and negative reviews (25,000 each). This balance is crucial for building an effective sentiment analysis model.
